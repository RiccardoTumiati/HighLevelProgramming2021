1\. Perform the following operations on plain `txt` files:

+ create a list of integrer numbers and then save it to a text file named "data_int.txt". Run the `cat` command to print the content of the file.
+ create a matrix of 5x5 floats and then save it to a text file named "data_float.txt". Use the `cat` command to print the content of the file.
+ load the txt file of the previous point and convert it to a csv file by hand.

```python
import csv

l=np.random.randint(low=0,high=100,size=5)
out_file_name = "data_int.txt"
with open(out_file_name, 'w') as outfile:
    outfile.write(np.array2string(l)+'\n')
!cat data_int.txt

m=np.random.rand(5,5)
out_file_name = "data_float.txt"
with open(out_file_name, 'w') as outfile:
    outfile.write(np.array2string(m)+'\n')
!cat data_float.txt

out_file_csv = "data_csv.csv"
with open(out_file_name, 'r') as infile:
    with open(out_file_csv, 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        for line in csv.reader(infile, delimiter=' '):
            while ('' in line):
                line.remove('')
            for i,j in zip(line,range(0,len(line))):
                line[j]=i.replace('[','').replace(']','')
            writer.writerow(line)
```

2\. Load the file *user_data.json*, which can be found at:

- https://www.dropbox.com/s/sz5klcdpckc39hd/user_data.json

and filter the data by the "CreditCardType" when it equals to "American Express". Than save the data to a new CSV file.

```python
!wget https://www.dropbox.com/s/sz5klcdpckc39hd/user_data.json
```

```python
import json
data = json.load(open('user_data.json'))

def AmericanExpress(x):
    return x["CreditCardType"]=="American Express"

result=list(filter(AmericanExpress,data))
fieldnames = ['ID', 'JobTitle', 'EmailAddress', 'FirstNameLastName','CreditCard','CreditCardType']
with open('output.csv', 'w',newline='') as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(result)

data = pd.read_csv('output.csv')
data
```
